```python
#%% [markdown]
# # Predição de função quadrática com uma Rede Neural (MLP) em NumPy
# Função alvo: y = 2x^2 - 3x + 7

#%%
import math
import numpy as np
import matplotlib.pyplot as plt


plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams["axes.grid"] = True

def f_quadratica(x):
    return 2*x**2 - 3*x + 7


rng = np.random.default_rng(42)
n_train = 200
n_test  = 200

X_train = rng.uniform(-3.0, 3.0, size=(n_train, 1))
y_train = f_quadratica(X_train)


y_train_noisy = y_train + rng.normal(0.0, 0.4, size=y_train.shape)

X_test  = rng.uniform(-3.0, 3.0, size=(n_test, 1))
y_test  = f_quadratica(X_test)


plt.scatter(X_train, y_train_noisy, s=12)
plt.title("Amostras de treino (com ruído)")
plt.xlabel("x"); plt.ylabel("y")
plt.show()

#%% [markdown]
# ## MLP 1→H→1 com ativação tanh (implementação do zero)

#%%
class MLP:
    def __init__(self, in_dim=1, hidden_dim=32, out_dim=1, seed=0):
        rng = np.random.default_rng(seed)
        # Inicialização de Glorot/Xavier para tanh
        lim1 = math.sqrt(6/(in_dim + hidden_dim))
        lim2 = math.sqrt(6/(hidden_dim + out_dim))
        self.W1 = rng.uniform(-lim1, lim1, size=(in_dim, hidden_dim))
        self.b1 = np.zeros((1, hidden_dim))
        self.W2 = rng.uniform(-lim2, lim2, size=(hidden_dim, out_dim))
        self.b2 = np.zeros((1, out_dim))

    def forward(self, X):
        Z1 = X @ self.W1 + self.b1          # (N,H)
        A1 = np.tanh(Z1)                     # (N,H)
        Z2 = A1 @ self.W2 + self.b2          # (N,1)
        return Z1, A1, Z2

    def predict(self, X):
        _, _, Z2 = self.forward(X)
        return Z2

    @staticmethod
    def mse(y_pred, y_true):
        return float(np.mean((y_pred - y_true)**2))

    def train(self, X, y, lr=5e-3, epochs=4000, batch_size=64, verbose=False):
        N = X.shape[0]
        losses = []
        for ep in range(epochs):
            idx = np.random.permutation(N)
            Xs, ys = X[idx], y[idx]

            for start in range(0, N, batch_size):
                end = start + batch_size
                xb = Xs[start:end]
                yb = ys[start:end]

                # Forward
                Z1, A1, Z2 = self.forward(xb)
                y_pred = Z2

                # dL/dy_pred (MSE)
                dY = (2.0 / yb.shape[0]) * (y_pred - yb)  # (B,1)

                # Camada 2
                dW2 = A1.T @ dY                            # (H,1)
                db2 = np.sum(dY, axis=0, keepdims=True)    # (1,1)

                # Camada 1
                dA1 = dY @ self.W2.T                       # (B,H)
                dZ1 = dA1 * (1 - np.tanh(Z1)**2)           # derivada tanh
                dW1 = xb.T @ dZ1                           # (1,H)
                db1 = np.sum(dZ1, axis=0, keepdims=True)   # (1,H)


                self.W2 -= lr * dW2
                self.b2 -= lr * db2
                self.W1 -= lr * dW1
                self.b1 -= lr * db1


            losses.append(self.mse(self.predict(X), y))
            if verbose and (ep+1) % 500 == 0:
                print(f"Época {ep+1}/{epochs} - MSE: {losses[-1]:.6f}")
        return losses

#%% [markdown]


mlp = MLP(in_dim=1, hidden_dim=32, out_dim=1, seed=7)
losses = mlp.train(X_train, y_train_noisy, lr=5e-3, epochs=4000, batch_size=64, verbose=False)

plt.plot(losses)
plt.title("Curva de aprendizado (MSE)")
plt.xlabel("Épocas"); plt.ylabel("MSE")
plt.show()

#%% [markdown]
# ## Avaliação e gráficos


x_grid = np.linspace(-3, 3, 400).reshape(-1,1)
y_true_grid = f_quadratica(x_grid)
y_pred_grid = mlp.predict(x_grid)

plt.scatter(X_train, y_train_noisy, s=12, label="Treino (ruído)")
plt.plot(x_grid, y_true_grid, label="Função real: 2x²-3x+7")
plt.plot(x_grid, y_pred_grid, label="MLP (previsão)")
plt.title("Ajuste da MLP à função quadrática")
plt.xlabel("x"); plt.ylabel("y")
plt.legend()
plt.show()

y_pred_test = mlp.predict(X_test)
mse_test = np.mean((y_pred_test - y_test)**2)
mae_test = np.mean(np.abs(y_pred_test - y_test))
print(f"MSE (teste): {mse_test:.6f}")
print(f"MAE (teste): {mae_test:.6f}")
```
